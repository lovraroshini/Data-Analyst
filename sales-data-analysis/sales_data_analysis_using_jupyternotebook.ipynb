```python
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])
df["Month"] = df["Sale_Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Sales_Amount"].sum()
monthly_sales
```
